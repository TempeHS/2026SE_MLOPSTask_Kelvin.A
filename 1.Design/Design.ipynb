# Design

## Defining the business problem to be solved

League of Legends is one of the worlds most popular competitive esports titles, boasting over millions globally in terms of viewership. Often coaches, and announcers may not realise what is truly happening inside a match, and may misjudge the direction the game is moving. Momentum shifts, macro / micro advantages, and subtle shifts in strategies frequently go unnoticed or are misread, leading to misinformed decisions or inconsistent narratives. This not only impacts viewer understanding and engagement. this reveals a significant opportunity in creating a ML system that takes in match data to predict outcomes before 15 minutes

## Refactoring the business problem into a machine learning problem
To develop a machine-learning model that predicts the outcome of a match as of/before 15 minutes into the game, using early game statistics
